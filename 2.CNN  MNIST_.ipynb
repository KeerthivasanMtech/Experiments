```
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
```

```
batch_size = 32
num_classes = 10
epochs = 10

# input image dimensions
img_rows, img_cols = 28, 28
```

```
# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

x_train = x_train.reshape((x_train.shape[0], 28, 28, 1))
x_test = x_test.reshape((x_test.shape[0], 28, 28, 1))

x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```
# convert class vectors to binary class matrices
from tensorflow.keras import utils
y_train = utils.to_categorical(y_train, num_classes)
y_test = utils.to_categorical(y_test, num_classes)
```

```
from keras import regularizers
regv = 0.0001

model = Sequential()
model.add(Conv2D(128, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1), kernel_regularizer=regularizers.l2(regv)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))  # Add dropout here

model.add(Conv2D(16, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(regv)))
model.add(Conv2D(34, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(regv)))
model.add(Conv2D(64, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(regv)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))  # Add dropout here

model.add(Flatten())
model.add(Dense(46, activation='relu', kernel_regularizer=regularizers.l2(regv)))
model.add(Dropout(0.5))  # Add dropout here

model.add(Dense(38, activation='relu', kernel_regularizer=regularizers.l2(regv)))
model.add(Dense(num_classes, activation='softmax', kernel_regularizer=regularizers.l2(regv)))

model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
```

```
model.summary()
```

```
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(x_train, y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(x_test, y_test))
```

```
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
